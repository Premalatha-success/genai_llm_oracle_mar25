```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/content/loan_prediction.csv")
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.describe(include="all")
```

```python
med_loanamount=df["LoanAmount"].median()
med_loanamount
```

```python
### Data Preparation-> Handle the missing values
df["LoanAmount"]=df["LoanAmount"].replace(np.nan,med_loanamount)
```

```python
df.isnull().sum()
```

```python
med_term=df["Loan_Amount_Term"].median()
med_term
```

```python
df["Loan_Amount_Term"]=df["Loan_Amount_Term"].replace(np.nan,med_term)
```

```python
med_history=df["Credit_History"].median()
med_history
```

```python
df["Credit_History"]=df["Credit_History"].replace(np.nan,med_history)
```

```python
df.isnull().sum()
```

```python
mode_gen=df["Gender"].mode()[0]
mode_gen
```

```python
df["Gender"]=df["Gender"].replace(np.nan,mode_gen)
```

```python
mode_marr=df["Married"].mode()[0]
df["Married"]=df["Married"].replace(np.nan,mode_marr)
```

```python
mode_dep=df["Dependents"].mode()[0]
df["Dependents"]=df["Dependents"].replace(np.nan,mode_dep)
```

```python
mode_emp=df["Self_Employed"].mode()[0]
df["Self_Employed"]=df["Self_Employed"].replace(np.nan,mode_emp)
```

```python
df.isnull().sum()
```

```python
df=df.drop("Loan_ID",axis=1)#axis=1 means column
```

```python
df.shape
```

```python
df.dtypes
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
for col in df.select_dtypes(include=['object']).columns:
    df[col] = le.fit_transform(df[col])
```

```python
df.head()
```

```python
df.dtypes
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression #classifier
```

```python
y=df[["Loan_Status"]]
X=df.drop("Loan_Status",axis=1)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3)
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
model=LogisticRegression()
model.fit(X_train,y_train)
```

```python
model.score(X_train,y_train)
```

```python
model.score(X_test,y_test)
```
